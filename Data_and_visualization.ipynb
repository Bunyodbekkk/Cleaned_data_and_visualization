```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
d=pd.read_html("https://en.tutiempo.net/climate/ws-384570.html",index_col=0)
df=d[3]
df=df[df.index>1990]
df=df.drop('PP',axis=1)
df=df.replace("-",np.nan)
df=df.astype(dtype='float')
df=df.fillna(method='ffill')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.lineplot(data=df)
plt.title("Toshkent shaxri yillik ob_havosi")
plt.xlabel("Yil")
plt.show()
```

```python
df=df.drop('SN',axis=1)
column=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",        
        "Yillik o'rtacha shamol tezligi",
        "Yomg'irli kunlar soni",
        "Bo'ronli kunlar soni",
        "Tumanli kunlar soni",
        "To'fonli kunlar soni",
        "Do'l yoqqan kunlar"]
df.columns=column
plt.figure(figsize=(16,6))
sns.lineplot(data=df)
plt.xlabel('Yil')
plt.ylabel('Harorat')
plt.show()
```

```python
labels=["Yillik o'rtacha harorat",
        "Yillik o'rtacha maksimal harorat",
        "Yillik o'rtacha minimal harorat",        
        # "Yillik o'rtacha shamol tezligi",
        # "Yomg'irli kunlar soni",
        # "Bo'ronli kunlar soni",
        # "Tumanli kunlar soni",
        # "To'fonli kunlar soni",
        # "Do'l yoqqan kunlar"
        ]
df1=df[labels]
plt.figure(figsize=(16,6))
sns.lineplot(data=df1, linewidth=2.0)
plt.title("Toshkent shahri uchun ob havo ko'rsatkichlari")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.legend(labels, loc=1)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df.index,y=df["Yillik o'rtacha shamol tezligi"])
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(30,5))
sns.heatmap(data=df[["Yomg'irli kunlar soni","Bo'ronli kunlar soni", "Tumanli kunlar soni","To'fonli kunlar soni","Do'l yoqqan kunlar"]].T,linewidths=1,cmap='YlGnBu',square=True)
```
